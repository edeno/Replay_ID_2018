```python
from loren_frank_data_processing import get_position_dataframe

position_info = get_position_dataframe(epoch_key, ANIMALS)
```

```python
position = position_info.loc[:, ['x_position', 'y_position']].values
```

```python
from loren_frank_data_processing.position import make_track_graph
from loren_frank_data_processing.track_segment_classification import (
    get_track_segments_from_graph, project_points_to_segment)

def get_projected_track_positions(track_graph, track_segment_id, position):
    track_segment_id[np.isnan(track_segment_id)] = 0
    track_segment_id = track_segment_id.astype(int)

    track_segments = get_track_segments_from_graph(track_graph)
    projected_track_positions = project_points_to_segment(
        track_segments, position)
    n_time = projected_track_positions.shape[0]
    return projected_track_positions[(
        np.arange(n_time), track_segment_id)]


track_segment_id = position_info.track_segment_id.values
track_graph, _ = make_track_graph(epoch_key, ANIMALS)

projected_position = get_projected_track_positions(track_graph, track_segment_id, position)
```

```python
import matplotlib.animation as animation
from loren_frank_data_processing.track_segment_classification import plot_track

# Set up formatting for the movie files
Writer = animation.writers['ffmpeg']
writer = Writer(fps=15, metadata=dict(artist='Me'), bitrate=1800)

fig = plt.figure(figsize=(7, 7))
plt.plot(position[:, 0], position[:, 1], color='lightgrey', zorder=-10)
plot_track(track_graph, plt.gca())

plt.xlim(position[:, 0].min() - 1, position[:, 0].max() + 1)
plt.ylim(position[:, 1].min() + 1, position[:, 1].max() + 1)
plt.xlabel('x-position')
plt.ylabel('y-position')
plt.title('Linearized vs. Actual Position')

actual_line, = plt.plot([], [], 'g-', label='actual position', linewidth=3, zorder=100)
predicted_line, = plt.plot([], [], 'b-', label='linearized position', linewidth=3, zorder=101)
head = plt.scatter([], [], s=80, zorder=102, color='b')
plt.legend()


def update_line(time_ind):
    start_ind = max(0, time_ind - 33)
    time_slice = slice(start_ind, time_ind)

    actual_line.set_data(position[time_slice, 0], position[time_slice, 1])
    predicted_line.set_data(projected_position[time_slice, 0],
                            projected_position[time_slice, 1])
    head.set_offsets(projected_position[time_ind])
    
    return actual_line, predicted_line


line_ani = animation.FuncAnimation(fig, update_line, frames=position.shape[0],
                                   interval=50, blit=True)
line_ani.save('actual_vs_linearized.mp4', writer=writer)
```
